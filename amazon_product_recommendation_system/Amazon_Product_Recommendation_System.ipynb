## **Section C: Content-Based Filtering**

A content-based recommendation system suggests items similar to those a user has liked in the past by analyzing item attributes (metadata like genre, director, description). It builds a user profile based on these features to predict preferences and recommend new, similar content. In our Amazon dataset, the best features to use are `product_name`, `category`, and `about_product`

#### **Vectorize product text (about_product + product_name) using TF-IDF**

The standard way to perform content-based filtering is by converting that text into numbers using TF-IDF (Term Frequency-Inverse Document Frequency) and then measuring how "close" products are using Cosine Similarity.

First, we will combine the descriptive columns into a single "metadata" string that the computer can analyze.

```python
# Remove duplicate product_id entries, keeping the first occurrence
amazon_unique_product_ids_df = amazon.drop_duplicates(subset='product_id', keep='first').reset_index(drop=True)

# Combine relevant text features into one column
amazon_unique_product_ids_df['metadata'] = amazon_unique_product_ids_df['product_name'] + ' ' + amazon_unique_product_ids_df['category_level_1'] + ' ' + amazon_unique_product_ids_df['category_level_2'] + ' ' + amazon_unique_product_ids_df['about_product']

original_count = len(amazon['product_id'])
unique_count = len(amazon_unique_product_ids_df['product_id'])
print(f"Removed {original_count - unique_count} duplicate products.")
print("_" *30)

amazon_unique_product_ids_df['metadata']
```

#### **Convert Text to Vectors (TF-IDF)**
Computers can't read, so we turn words into a mathematical matrix where unique/important words (like "Noise-Cancelling") get higher weights than common words (like "the").

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the vectorizer (removing common English "stop words")
tfidf = TfidfVectorizer(stop_words='english')

# Construct the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(amazon_unique_product_ids_df['metadata'])

print(f"TF-IDF matrix shape: {tfidf_matrix.shape}")
```

#### **Vectorize product text (about_product + product_name) using Embeddings**

**Embeddings** are the modern "gold standard" for recommendation systems, often outperforming TF-IDF because they understand semantics (meaning) rather than just matching keywords.

While TF-IDF treats words as isolated tokens, Embeddings represent them as dense vectors in a high-dimensional space.

**Why Embeddings are "Smarter"**

If one product description uses the word "Cinema" and another uses "Movie," TF-IDF sees them as completely different. An Embedding model knows they are mathematically similar.

We will use a pre-trained model like Sentence-Transformers (built on BERT) to turn your Amazon product descriptions into 384 or 768-dimensional vectors.

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import warnings
from transformers import logging

# Suppress the 'UNEXPECTED' / 'MISSING' library warnings to keep output clean
logging.set_verbosity_error()
warnings.filterwarnings('ignore', category=UserWarning)

# Load a pre-trained model (Semantic meaning of text)
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Encode metadata column into vectors
# This creates a 'feature map' representing the semantic meaning of products
embeddings = model.encode(amazon_unique_product_ids_df['metadata'].tolist(), show_progress_bar=True)
```

#### **Calculate Similarity Scores**
Now, we calculate the Cosine Similarity. This creates a "similarity map" where every product is compared to every other product. A score of 1.0 means they are identical; 0.0 means they share no keywords.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the similarity scores for TF-IDF
tf_idf_cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

tf_idf_cosine_sim
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the similarity scores for Embeddings
embeddings_cosine_sim = cosine_similarity(embeddings, embeddings)

embeddings_cosine_sim
```

#### **Recommendation Function**
This function takes a product_id, finds its index, and pulls the top 5 most similar items.

```python
def get_recommendations(product_id, cosine_sim, get_hybrid_score=False):
  # Get the index and metadata for the base product
  base_product = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'] == product_id].iloc[0]
  idx = base_product.name
  base_subcat = base_product['category_level_2']

  # Get pairwise similarity scores
  sim_scores = list(enumerate(cosine_sim[idx]))

  # Apply Hybrid Scoring if requested
  if get_hybrid_score:
    weighted_sim_scores = [(x, y * 0.7) for x, y in sim_scores]
    amazon['norm_rating'] = (amazon['rating'] - min_rating) / (max_rating - min_rating)
    weighted_norm_ratings = [(x, amazon.loc[x, 'norm_rating'] * 0.3) for x, y in sim_scores]
    sim_scores = [(s[0], s[1] + r[1]) for s, r in zip(weighted_sim_scores, weighted_norm_ratings)]

  # Sort and get top 5 (excluding itself)
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  # Get the scores of the 5 most similar products (excluding itself)
  sim_scores = sim_scores[1:6]

  # Get the product indices
  final_indices = [i[0] for i in sim_scores]

  return amazon_unique_product_ids_df[['product_name', 'category_level_1', 'category_level_2', 'actual_price', 'rating']].loc[final_indices]
```

#### **Add category, price, and discount to enhance content vectors**

To create a truly powerful "Hybrid" recommendation system, we will combine our high-dimensional content vectors (text) with our metadata features (category, price, and discount).

This prevents the system from recommending a $1,000 product when the user is looking at a $20 "budget" item, even if the descriptions are similar.

#### **1. Pre-processing Metadata**
Numerical data must be scaled so that large numbers (like a $5,000 price) don't overpower smaller numbers (like a 4.5 rating).

* Category: Convert to "One-Hot Encodings" (0s and 1s).

* Price & Discount: Use Min-Max Scaling to bring them into a 0–1 range.

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Convert category columns to one-hot encoding
encoder = OneHotEncoder(sparse_output=False)
encoded_categories = encoder.fit_transform(amazon_unique_product_ids_df[['category_level_1', 'category_level_2']])

# Scale numerical features
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(amazon_unique_product_ids_df[['actual_price', 'discount_percentage', 'rating']])

encoded_categories.shape, scaled_features.shape
```

#### **2. Combining Vectors (Feature Fusion)**
We can connect our text embeddings and numerical features together using np.hstack. We can multiply specific features by a weight if we want them to be more influential.

```python
# Increase the weight of Category and Price to ensure relevance
# We multiply categorical features by 5 to force recommendations within the same category
weighted_categories = encoded_categories * 5.0

# We give Price a weight of 2.0 to keep it in a similar bracket
scaled_features[:, 0] *= 2.0

# Combine everything into one "Master Vector"
# Text embeddings (384 dims) + Weighted Categories + Scaled Numerical Features
combined_vectors = np.hstack((embeddings, weighted_categories, scaled_features))

print(f"New combined vector shape: {combined_vectors.shape}")
```

#### **4. Refining the Search**
Now that the vectors are combined, we calculate Cosine Similarity just like before. The difference is that now the "distance" between products accounts for their price bracket and category.

```python
# Re-calculate similarity with the new weighted vectors

# Check for NaNs and fill them with 0 to prevent the ValueError
if np.isnan(combined_vectors).any():
    combined_vectors = np.nan_to_num(combined_vectors)

# Calculate cosine similarity for combined vectors
combined_vectors_cosine_sim = cosine_similarity(combined_vectors, combined_vectors)

print("Similarity matrix updated with categorical weighting.\n")
combined_vectors_cosine_sim
```

#### **Evaluate recommendations:**
How diverse and relevant are the content-based results?

#### **1. TF-IDF (Term Frequency-Inverse Document Frequency)**
* **Relevance**: High keyword matching. It is excellent at finding products with identical technical specifications (e.g., matching '65W', 'Braided', 'Type C').

* **Diversity**: Low. It tends to stay very close to the specific terminology used in the base product, often missing similar items that use different synonyms.

#### **2. Semantic Embeddings (all-MiniLM-L6-v2)**
* **Relevance**: Good. It understands context (e.g., recognizing that 'Cinema' and 'Movie' are related). It is less prone to failing if a keyword is missing but the intent is the same.

* **Diversity**: Higher. It pulls in a broader range of products that are conceptually similar, though sometimes it can be too broad, occasionally pulling in items from different price brackets or sub-categories if the descriptions are stylistically similar.

#### **3. Combined Vectors (Feature Fusion)**
* **Relevance**: Excellent. By weighting categories and scaling prices, this method ensures that a 'Budget' heater doesn't get a 'Premium' AC unit as a recommendation just because both mention 'heating/cooling'.

* **Diversity**: Balanced. It provides the most "logical" recommendations for a real-world store, maintaining semantic similarity while respecting product hierarchy and user expectations regarding price.

#### **Summary Evaluation**
For this Amazon dataset, **Combined Vectors** provide the most robust recommendations. While **Embeddings** are superior for capturing human-like meaning, the addition of hard metadata (Price & Category) in the Combined approach is essential to prevent 'logical' errors (like price-mismatches). **TF-IDF** remains a useful baseline for strict technical attribute matching but lacks the 'intelligence' needed for modern discovery.

## **Section D: Collaborative Filtering (User–Item) :**
#### Create a user-item matrix using user_id, product_id, and rating.

```python
# Create a user-item matrix
# We use pivot_table to handle cases where there might be duplicate entries (though we've cleaned many)
user_item_matrix = amazon.pivot_table(index='user_id', columns='product_id', values='rating')

print(f"User-Item Matrix Shape: {user_item_matrix.shape}")
print(f"Number of Users: {user_item_matrix.shape[0]}")
print(f"Number of Products: {user_item_matrix.shape[1]}")

# Display a snippet of the matrix
display(user_item_matrix.head())

# Calculate Sparsity
total_elements = user_item_matrix.size
non_zero_elements = user_item_matrix.notna().sum().sum()
sparsity = (1 - (non_zero_elements / total_elements)) * 100
print(f"Matrix Sparsity: {round(sparsity, 2)}%")
```

#### User-User Collaborative Filtering (cosine similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity

# Fill NaN values with 0 to perform matrix operations
user_item_matrix_filled = user_item_matrix.fillna(0)

# Calculate User-User Similarity
user_sim = cosine_similarity(user_item_matrix_filled)
user_sim_df = pd.DataFrame(user_sim, index=user_item_matrix.index, columns=user_item_matrix.index)

print(f"User Similarity Matrix Shape: {user_sim_df.shape}")
# Comparison Note:
# Given the 99.92% sparsity, Matrix Factorization (SVD) would likely outperform
# simple Cosine Similarity by uncovering latent patterns rather than relying
# on direct overlaps between sparse user vectors.
```

```python
def get_user_user_recommendations(user_id, num_recs=5):
    if user_id not in user_item_matrix.index:
        return "User ID not found in the matrix."

    # Get similarity scores for the target user
    similar_users = user_sim_df[user_id].sort_values(ascending=False)[1:11]

    # Get products highly rated by similar users but not yet rated by the target user
    user_rated_products = user_item_matrix.loc[user_id].dropna().index

    similar_users_ratings = user_item_matrix.loc[similar_users.index].mean().sort_values(ascending=False)

    recs = similar_users_ratings.drop(labels=user_rated_products).head(num_recs)

    return amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'].isin(recs.index)][['product_name', 'category_level_1', 'rating']]

# Test with a sample user from the matrix
sample_user = user_item_matrix.index[0]
print(f"Recommendations for user: {sample_user}")
display(get_user_user_recommendations(sample_user))
```

#### Item-Item Collaborative Filtering (cosine or Pearson)

Unlike user-user filtering, this approach looks for products that are similar based on how users have rated them.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Transpose the user-item matrix to get item-user matrix
item_user_matrix = user_item_matrix.T.fillna(0)

# Calculate Item-Item Similarity
item_sim = cosine_similarity(item_user_matrix)
item_sim_df = pd.DataFrame(item_sim, index=item_user_matrix.index, columns=item_user_matrix.index)

print(f"Item Similarity Matrix Shape: {item_sim_df.shape}")
display(item_sim_df.head())
```

```python
def get_item_item_recommendations(product_id, num_recs=5):
    if product_id not in item_sim_df.index:
        return "Product ID not found in the matrix."

    # Get similarity scores for the target product
    similar_items = item_sim_df[product_id].sort_values(ascending=False)[1:num_recs+1]

    # Get metadata for the recommended products
    recs = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'].isin(similar_items.index)]

    return recs[['product_name', 'category_level_1', 'rating', 'actual_price']]

# Test with a sample product from the matrix
sample_product = item_sim_df.index[0]
sample_product_name = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'].isin([sample_product])]['product_name']
sample_product_name.columns = ['']
print('Product id =>', sample_product)
print('Product name =>', sample_product_name.values[0])
print('\n')
print("Recommendations for this product")
display(get_item_item_recommendations(sample_product))
```

**Item Similarity Matrix**: The user-item matrix is transposed to calculate similarities between products. This allows the system to recommend items that 'people who liked this also liked.'

**Functionality**: The get_item_item_recommendations function identifies the top 5 most similar products based on rating overlaps.

**Test Results**: For the sample product (a Computer Accessory), the system suggested related tech items like a wireless mouse and charging cables, as well as highly-rated household appliances like a vacuum cleaner and multi-cooker, reflecting broader purchasing patterns in the dataset.

#### Comparison of User-Based and Item-Based Collaborative Filtering for this dataset:

**1. User-User Collaborative Filtering**

* **How it works**: Finds users with similar rating histories and suggests items they liked which the target user hasn't seen.
* **Observations**: In this dataset, user-user filtering produced recommendations across different categories (e.g., suggesting an Air Fryer to a user who bought a phone). This is great for discovery and serendipity.

* **Challenges**: Because our matrix is 99.92% sparse, it is very hard to find users who have rated the exact same products, often leading to less reliable similarity scores.

**2. Item-Item Collaborative Filtering**

* **How it works**: Finds products that are commonly rated together by the same users.

* **Observations**: This produced highly relevant technical matches (e.g., suggesting a mouse and cables for a USB adapter). It is generally more stable because product relationships don't change as quickly as user preferences.

* **Efficiency**: Usually preferred for large e-commerce sites like Amazon because there are often more ratings per item than per user, making the similarity scores more robust.

**Summary**

For this specific case, Item-Item filtering appeared to give more logically consistent results, whereas User-User filtering provided broader, more exploratory suggestions.

#### **Section E: Hybrid Recommender (Content + Collaborative):**

#### Design a hybrid strategy:
Score fusion: 0.6 * CF_score + 0.4 * Content_score

Let's select a sample product ID, extract its similarity scores from both the content-based (combined_vectors_cosine_sim) and collaborative (item_sim_df) matrices, and then align and normalize them to ensure they are on the same scale for the hybrid recommender.



```python
from sklearn.preprocessing import MinMaxScaler

# 1. Select a sample product ID
sample_pid = 'B002PD61Y4'

# 2. Extract content-based similarity scores
# Find the integer index of the product in the unique products dataframe
idx = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'] == sample_pid].index[0]
content_scores = combined_vectors_cosine_sim[idx]

# 3. Extract collaborative filtering similarity scores
# These are stored in item_sim_df where columns/index are product_ids
cf_scores_series = item_sim_df[sample_pid]

# 4. Align CF scores to the order of products in amazon_unique_product_ids_df
# This ensures the i-th element in both arrays refers to the same product
cf_scores_aligned = cf_scores_series.reindex(amazon_unique_product_ids_df['product_id']).fillna(0).values

# 5. Normalize both sets of scores to [0, 1] range
scaler = MinMaxScaler()

# Reshape for scaler and transform
content_scores_norm = scaler.fit_transform(content_scores.reshape(-1, 1)).flatten()
cf_scores_norm = scaler.fit_transform(cf_scores_aligned.reshape(-1, 1)).flatten()

print(f'Similarity scores prepared for Product ID: {sample_pid}')
print(f'Content Scores - Min: {content_scores_norm.min():.4f}, Max: {content_scores_norm.max():.4f}')
print(f'CF Scores      - Min: {cf_scores_norm.min():.4f}, Max: {cf_scores_norm.max():.4f}')
print(f'Aligned lengths: Content={len(content_scores_norm)}, CF={len(cf_scores_norm)}')
```

```python
from sklearn.preprocessing import MinMaxScaler

def get_hybrid_item_recommendations(product_id, content_sim_matrix, cf_sim_df):
    # 1. Get the integer index of the input product_id
    if product_id not in amazon_unique_product_ids_df['product_id'].values:
        return "Product ID not found in unique products list."

    idx = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'] == product_id].index[0]

    # 2. Extract content similarity scores
    content_scores = content_sim_matrix[idx]

    # 3. Extract and align collaborative similarity scores
    if product_id in cf_sim_df.index:
        cf_scores_series = cf_sim_df[product_id]
        cf_scores_aligned = cf_scores_series.reindex(amazon_unique_product_ids_df['product_id']).fillna(0).values
    else:
        cf_scores_aligned = np.zeros(len(amazon_unique_product_ids_df))

    # 4. Normalize scores to [0, 1] range
    scaler = MinMaxScaler()
    content_norm = scaler.fit_transform(content_scores.reshape(-1, 1)).flatten()
    cf_norm = scaler.fit_transform(cf_scores_aligned.reshape(-1, 1)).flatten()

    # 5. Calculate Hybrid Score (40% Content, 60% Collaborative)
    hybrid_scores = (0.4 * content_norm) + (0.6 * cf_norm)

    # 6. Store scores in a copy of the dataframe
    recommendations_df = amazon_unique_product_ids_df.copy()
    recommendations_df['hybrid_score'] = hybrid_scores

    return recommendations_df

print('Function get_hybrid_item_recommendations defined successfully.')
```

```python
# 1. Define a sample product ID to test
sample_pid = 'B002PD61Y4'
sample_product_name = amazon_unique_product_ids_df[amazon_unique_product_ids_df['product_id'].isin([sample_pid])]['product_name']
sample_product_name.columns = ['']
sample_product_name = sample_product_name.values[0]

# 2. Call the hybrid recommendation function
hybrid_recs_df = get_hybrid_item_recommendations(sample_pid, combined_vectors_cosine_sim, item_sim_df)

# 3. Exclude the input product itself
filtered_recs = hybrid_recs_df[hybrid_recs_df['product_id'] != sample_pid]

# 4. Sort by hybrid_score in descending order
sorted_recs = filtered_recs.sort_values(by='hybrid_score', ascending=False)

# 5. Select the top 5 rows
top_5_hybrid = sorted_recs.head(5)

# 6. Display the recommendations with metadata
print('Top 5 Hybrid Recommendations for:\n')
print(f"Product id => {sample_pid}")
print(f"Product name => {sample_product_name}")
print('_' * 100)
display(top_5_hybrid[['product_name', 'category_level_1', 'actual_price', 'rating', 'hybrid_score']])
```

### Hybrid Strategy Summary

**How does the hybrid approach balance semantic product relevance with actual user purchasing behavior patterns?**
The hybrid approach balances these two signals by applying a weighted average of their normalized scores. Semantic relevance is captured through **Content-Based Filtering (40% weight)**, which uses product embeddings and metadata to ensure items are functionally similar (e.g., matching a WiFi adapter with other networking tools). User purchasing patterns are captured through **Collaborative Filtering (60% weight)**, which prioritizes products that real customers frequently interact with or buy together. This 40/60 split ensures that while the system respects the physical characteristics of the product, it leans more heavily on actual market behavior and community preferences.

### Data Analysis Key Findings

*   **Normalization Strategy:** Using `MinMaxScaler`, both content and collaborative scores were scaled to a strict **\[0.0, 1.0\] range**. This step was critical to prevent one model's score magnitude from dominating the other before the weighting was applied.
*   **Weighted Scoring:** The implementation used a ratio of **40% content-based similarity** and **60% collaborative similarity**, favoring historical user interaction data while maintaining category relevance.
*   **Top Recommendations Performance:** For a sample product (Wi-Fi Adapter `B002PD61Y4`), the system generated top 5 recommendations including brands like Skadioo, D-Link, and Gizga.
*   **Score Stability:** The top hybrid scores were clustered around **0.39**, indicating a strong consensus between the two recommendation engines for the top-ranked items.

### Insights or Next Steps

*   **Weight Optimization:** Future iterations could involve A/B testing different weight ratios (e.g., 50/50 or 30/70) to see which combination leads to higher click-through rates or actual conversions in a production environment.

#### **Suggest how to improve hybrid performance further using real-world constraints like:**
* Popularity
* Recent purchases

#### **Enhancing Hybrid Recommendations with Real-World Constraints**

To optimize the recommendation engine for a production environment, we can incorporate the following three layers into our scoring logic:

**1. Popularity Bias (Bayesian Averaging)**
*   **The Problem:** High similarity scores might occasionally surface 'niche' or low-rated products.
*   **The Fix:** Add a **Popularity Multiplier**. Instead of using raw ratings, use a Bayesian Average that accounts for the `rating_count`. This ensures that a 5.0-star product with 2 reviews doesn't outrank a 4.8-star product with 20,000 reviews.

**2. Recency & Temporal Decay**
*   **The Problem:** Collaborative filtering might suggest products based on old trends (e.g., last year's smartphone model).
*   **The Fix:** Implement a **Time-Decay Function**. Multiply the similarity score by a factor like $e^{-\lambda t}$, where $t$ is the time since the product's last review or purchase. This keeps the recommendations 'fresh' and aligned with current market trends.

**3. Availability & Business Constraints**
*   **The Problem:** Recommending out-of-stock items leads to a poor user experience and lost conversion.
*   **The Fix:** Implement a **Hard Constraint Filter**. Before the final sort, join the recommendation list with an inventory status table. If `is_in_stock == False`, the product is either removed or moved to a 'Pre-order' section with a penalty to its hybrid score.

**Revised Hybrid Score Formula Example:**
$$Final\ Score = (w_1 \cdot Content + w_2 \cdot Collaborative) \times Recency\ Factor \times Popularity\ Weight$$

#### *Which model works best for new users?*

### **Recommendation Strategy for New Users (Cold-Start)**

#### **Winner: Hybrid Content-Collaborative Model**

For new users who have just joined the platform and have zero or very few reviews, **Collaborative Filtering** is ineffective because there is no historical data to calculate similarities with other users.

**Why the Hybrid Model is Best:**
1.  **Immediate Relevance:** As soon as a new user clicks on a single product, the Hybrid model uses the **Content-Based (40%)** component to find items with similar technical specs, categories, and price points.
2.  **Market Guidance:** It still applies the **Collaborative (60%)** logic to the *seed item* itself. This means the user sees products that are not just similar, but are also confirmed 'best-sellers' among the wider community.
3.  **Path to Personalization:** As the user continues to interact, the system dynamically shifts from 'Item-Similarity' (Content) to 'User-Preference' (Collaborative) as their profile grows.

**Business Impact:** This prevents the 'Empty State' problem, increasing the likelihood of the first purchase and higher user retention.

#### *Which model works best for returning users?*

### **Recommendation Strategy for Returning Users**

#### **Winner: Item-Item Collaborative Filtering**

For users with an existing history of interactions, **Item-Item Collaborative Filtering** (which we implemented using Cosine Similarity) is the most effective tool currently in our pipeline.

**Why Item-Item Filtering is Best for Returning Users:**
1.  **Stable Product Relationships:** Unlike user preferences which can be flighty, the relationship between products (e.g., people who buy a specific DSLR often buy a specific tripod) is statistically stable and highly reliable for cross-selling.
2.  **Overcoming Sparsity:** While User-User filtering struggles with our 99.92% sparsity, items generally have more ratings than individual users, making the item-similarity scores more robust and less likely to result in 'empty' recommendations.
3.  **Real-World Context:** It allows for the 'Frequently Bought Together' functionality, which is a major driver of Average Order Value (AOV) on platforms like Amazon.

**Business Impact:** High-quality item-to-item recommendations for returning users directly improve **Customer Lifetime Value (CLV)** by surfacing relevant accessories and upgrades based on their previous purchases.

#### *How can we recommend products with no ratings?*

### **How to Recommend Products with No Ratings**

#### **Strategy: Content-Based Fallback**

Products with no ratings suffer from the **Item Cold-Start** problem. Since no user-item interactions exist, similarity cannot be calculated via Collaborative Filtering.

**The Solution:**
1.  **Semantic Mapping**: Use the **Content-Based Model** (TF-IDF or Embeddings) to convert the new product's metadata (name, description, category) into a vector.
2.  **Attribute Matching**: Calculate the Cosine Similarity between this new vector and all existing products in the catalog.
3.  **Cross-Promotion**: Surface the new product on the pages of highly similar, established products under a "Similar Items" or "New Arrivals You Might Like" section.
4.  **Initial Discovery**: Use the user's past interest in specific categories or keywords to push the new product to their feed, effectively using the content features as a proxy for potential interest.

**Business Impact**: This ensures that new inventory gets immediate visibility, allowing it to collect its first ratings and eventually transition into the more powerful Collaborative Filtering models.

#### *What KPIs should Amazon track to measure success?*

### **KPIs to Measure Recommendation Success**

To evaluate the performance of the hybrid recommendation system, the following Key Performance Indicators (KPIs) should be tracked:

#### **1. Business Metrics (User Impact)**
*   **Click-Through Rate (CTR)**: The percentage of users who click on a recommended product. This measures the immediate relevance of the suggestions.
*   **Conversion Rate (CVR)**: The percentage of clicks on recommendations that lead to an actual purchase. This is the ultimate measure of business value.
*   **Average Order Value (AOV)**: By suggesting relevant accessories (cross-selling), a good system should increase the total amount spent per transaction.
*   **Customer Lifetime Value (CLV)**: Tracking whether personalized recommendations lead to higher retention and repeat purchases over time.

#### **2. Technical Metrics (Model Accuracy)**
*   **Root Mean Square Error (RMSE)**: Measures the difference between predicted ratings and actual ratings (primarily for collaborative filtering/SVD models).
*   **Precision @ K**: Of the top 'K' recommendations shown to a user, how many were actually relevant or interacted with?
*   **Recall @ K**: How many of the items the user eventually bought were actually captured in the top 'K' recommendations?

#### **3. Catalog & Diversity Metrics**
*   **Catalog Coverage**: The percentage of the total product catalog that is actually being recommended. A system with low coverage only recommends 'blockbusters,' missing niche sales.
*   **Serendipity**: A measure of how 'surprising' or 'novel' the recommendations are, ensuring the user isn't just seeing the same brands repeatedly.
*   **Churn Rate of Recommendations**: How often the recommendation list refreshes to keep the user experience feeling new.
